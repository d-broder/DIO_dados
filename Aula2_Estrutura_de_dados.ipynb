**Listas**

```python
#Crindo uma lista de animais
animais = [1,2,3]
animais
```

```python
animais = ["cachorro", "gato", 12, 6.5]
animais
```

```python
#Imprimindo o primeiro elemento da lista
animais[0]
```

```python
#Imprimindo o quarto elemento da lista (i = 4 - 1)
animais[3]
```

```python
#Substituindo o primeiro elemento
animais[0] = "papagaio"
animais
```

```python
#Removendo "gato"
animais.remove("gato")
animais
```

```python
#Obtendo o tamanho da lista
len(animais)
```

```python
#Existe "gato" em lista?
"gato" in animais
```

```python
lista = [500, 30, 300, 80, 10]
```

```python
max(lista)
```

```python
min(lista)
```

```python
#Adicionar um elemento
animais.append("leão")
animais
```

```python
#Adicionar lista na lista
animais.append(["cachorro","coelho"])
animais
```

```python
#Adicionar mais de um elemento
animais.extend(["cobra", 6])
animais
```

```python
#Existem 0 "cachorro" pois está dentro de uma lista aninhada
animais.count("cachorro")
```

```python
animais.count("papagaio")
```

```python
lista.sort()
lista
```

```python
lista.sort(reverse=True)
lista
```
